```python
df["Weekend"] = df["Weekend"].astype(float)
df["Revenue"] = df["Revenue"].astype(float)
df
```

```python
df.loc[df["Revenue"]==1]
```

```python
df["Month"] = df["Month"].replace({'Feb': 2, "Mar":3, "May":5, "June":6, "Jul":7, "Aug":8, "Sep":9, "Oct":10, "Nov":11, "Dec":12})
```

month_list = []
for value in df["Month"]:
  if value == "Feb":
    month_list.append(2)
  elif value == "Mar":
    month_list.append(3)
  elif value == "May":
    month_list.append(5)
  elif value == "June":
    month_list.append(6)
  elif value == "Jul":

  elif value == "Aug":

  elif value == "Sep":

  elif value == "Oct":
    month_list.append(10)
  elif value == "Nov":
    month_list.append(11)
  elif value == "Dec":
    month_list.append(12)
  
df["Month"] = month_list

df


```python
#df = df.rename(columns = {"VisitorType_Other": "Other_VisitorType"})

df["VisitorType"] = df["VisitorType"].replace({'Other':'Different'})
df
```

```python
df = pd.get_dummies(df, drop_first=True).astype(float)
df
```
